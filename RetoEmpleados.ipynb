```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
import numpy as np
from google.colab import drive
from sklearn.preprocessing import LabelBinarizer
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
EmpleadosAttrition = pd.read_csv("/content/drive/MyDrive/diplomado/ingecaracteristicas/empleadosRETO.csv")
EmpleadosAttrition.head()
```

```python
EmpleadosAttrition= EmpleadosAttrition.drop(['EmployeeCount', 'Over18', 'StandardHours','EmployeeNumber'], axis=1)
EmpleadosAttrition.head()
```

```python
EmpleadosAttrition['HiringDate']
```

```python
EmpleadosAttrition['Year'] = EmpleadosAttrition['HiringDate'].str.split(pat='/').str[2].astype(int)
EmpleadosAttrition.head()
```

```python
EmpleadosAttrition['YearsAtCompany '] = 2018-EmpleadosAttrition['Year']
EmpleadosAttrition.head()
```

```python
EmpleadosAttrition=EmpleadosAttrition.rename(columns={'DistanceFromHome':'DistanceFromHome_km'})
EmpleadosAttrition['DistanceFromHome_km']
```

```python
EmpleadosAttrition['DistanceFromHome'] = EmpleadosAttrition['DistanceFromHome_km'].str.replace('km', '').astype(int)
EmpleadosAttrition['DistanceFromHome']
```

```python
EmpleadosAttrition= EmpleadosAttrition.drop(['Year', 'HiringDate', 'DistanceFromHome_km'], axis=1)
EmpleadosAttrition.head()
```

```python
SueldoPromedioDepto = EmpleadosAttrition.groupby('Department')['MonthlyIncome'].mean().reset_index()
SueldoPromedioDepto.rename(columns={'MonthlyIncome': 'SueldoPromedio'}, inplace=True)
SueldoPromedioDepto.head()
```

```python
EmpleadosAttrition['MonthlyIncome'] = (EmpleadosAttrition['MonthlyIncome'] - min(EmpleadosAttrition['MonthlyIncome'])) /(max(EmpleadosAttrition['MonthlyIncome']) - min(EmpleadosAttrition['MonthlyIncome']))
EmpleadosAttrition['MonthlyIncome']
```

```python

categoricas = ['BusinessTravel', 'Department', 'EducationField', 'JobRole', 'MaritalStatus', ]
lb = LabelBinarizer()
EmpleadosAttritiondummies = pd.get_dummies(EmpleadosAttrition[categoricas])
EmpleadosAttritiondummies = EmpleadosAttritiondummies.astype(int)
EmpleadosAttrition1 = EmpleadosAttrition.drop(categoricas, axis=1)
EmpleadosAttrition = pd.concat([EmpleadosAttrition1, EmpleadosAttritiondummies], axis=1)
EmpleadosAttrition['Attrition'] = lb.fit_transform(EmpleadosAttrition['Attrition'])
EmpleadosAttrition['Gender'] = lb.fit_transform(EmpleadosAttrition['Gender'])
EmpleadosAttrition['OverTime'] = lb.fit_transform(EmpleadosAttrition['OverTime'])
EmpleadosAttrition.head()
```

```python
EmpleadosAttrition.columns
```

```python
correlaciones = {}#diccionario

for columna in EmpleadosAttrition.columns:
    if columna != 'Attrition':
        correlacion = EmpleadosAttrition['Attrition'].corr(EmpleadosAttrition[columna])
        correlaciones[columna] = abs(correlacion)
corr = pd.DataFrame(correlaciones.items(), columns=['Variable','Correlación'])
corr
```

```python
menores = corr[corr['Correlación']< 0.1]['Variable'].tolist()
menores
```

```python
EmpleadosAttritionFinal=EmpleadosAttrition.drop(menores, axis=1)
EmpleadosAttritionFinal.head()
```

```python

pca = PCA()
EmpleadosAttritionPCA = pca.fit_transform( EmpleadosAttritionFinal)
print(EmpleadosAttritionPCA[:, 0])
```

```python
#varianzaex = pca.explained_variance_
varianzaexra = pca.explained_variance_ratio_
#puntos = pca.transform(X)
varianzafin = np.cumsum(varianzaexra)
num_componentes = np.argmax(varianzafin >= 0.8) + 1
componentes_df = pd.DataFrame(EmpleadosAttritionPCA[:, :num_componentes], columns=[f'C{i}' for i in range(num_componentes)])
EmpleadosAttritionFinal = EmpleadosAttritionFinal.assign(**componentes_df)
EmpleadosAttritionFinal.head()
```
